```python
pos = pd.read_csv('positive_bay_v3.csv',encoding='utf8')
rules = [
    ('"', ''),
    ("'", '')
]

pos_data = []

for i in pos['texts']:
    for rule in rules:#+text_norm:
        i = re.sub(rule[0],rule[1],i).strip()
    pos_data += [x.strip() for x in i.split('/')]
# for i in pos_data:
#     print(i)
print(len(pos_data))
```

```python
pos=pd.DataFrame({'texts':pos_data}).sample(frac=1).reset_index(drop=True)
pos['labels'] = 'positive'
pos['texts'] = pos['texts'].apply(lambda x:re.sub(r'[!?."]','',' '.join([i.strip() for i in ' '.join(x.split(',')).split('/') if i.strip()!=''])).lower())
pos['len'] = pos['texts'].apply(len)
pos.head(20)
```

```python
neg = pd.read_csv('negative_bay_v3.csv',encoding='utf8',header=None)
neg = neg.rename(columns={0: "texts"}).sample(frac=1).reset_index(drop=True)
neg['labels'] = 'negative'
neg['texts'] = neg['texts'].apply(lambda x:re.sub(r'[!?."]','',' '.join([i.strip() for i in ' '.join(x.split(',')).split('/') if i.strip()!=''])).lower())
neg['len'] = neg['texts'].apply(len)
neg.head(20)
```

```python
texts = []
with open('neutral_bay_v3.txt','r', encoding='utf8') as f:
    for line in f:
        line = line.strip()
        texts.append(line)
neu = pd.DataFrame({'texts': texts})
```

```python
# neu = pd.read_csv('neutral_bay_v3.csv',encoding='utf8',header=None)
neu = neu.sample(frac=1).reset_index(drop=True)
neu['labels'] = 'neutral'
neu['texts'] = neu['texts'].apply(lambda x:re.sub(r'[!?."]','',' '.join([i.strip() for i in ' '.join(x.split(',')).split('/') if i.strip()!=''])).lower())
neu['len'] = neu['texts'].apply(len)
neu.head(20)
```

```python
texts = []
with open('real_negative_v3.txt','r', encoding='utf8') as f:
    for line in f:
        line = line.strip()
        tmp = []
        for i in line.split(','):
            tmp += i.split('/')
        
        for i in tmp:
            texts.append(i)
neg2 = pd.DataFrame({'texts': texts})
```

```python
data = []
for _,i in neg2.iterrows():
    tmp = []
#     print(i)
    for j in [x.strip() for x in i['texts'].split(' ') if x.strip() != '']:
        tmp.append(j)
        if len(' '.join(tmp)) > 100:
            data.append(' '.join(tmp))
            tmp = []
    if 0< len(' '.join(tmp)) <= 100:
        data.append(' '.join(tmp))
neg2 = pd.DataFrame({'texts': data})
```

```python
neg2 = neg2.sample(frac=1).reset_index(drop=True)
neg2['labels'] = 'negative'
neg2['texts'] = neg2['texts'].apply(lambda x:re.sub(r'[!?."]','',' '.join([i.strip() for i in ' '.join(x.split(',')).split('/') if i.strip()!=''])).lower())
neg2['len'] = neg2['texts'].apply(len)
neg2.head(20)
```
